```python
import pandas as pd

def split_csv(csvFilePath, newDirPath, splitKey1, splitKey2):
    df = pd.read_csv(csvFilePath)
    for (n), group in df.groupby([splitKey1, splitKey2]):
        group.to_csv(f'{newDirPath}{n}.csv')

# https://stackoverflow.com/questions/419163/what-does-if-name-main-do
if __name__ == "__main__":
    split_csv(
        csvFilePath='./data/dataDC.csv', # CHANGE ME
        newDirPath='/data', # CHANGE ME
        splitKey1='state',
        splitKey2='session'# CHANGE ME
    )
```
